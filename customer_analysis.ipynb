# Customer Churn Analysis

This notebook analyzes customer churn data using Python, Pandas, Seaborn, and Matplotlib. The goal is to understand customer behavior through data visualization and identify patterns related to customer churn.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv("customer_churn.csv")

print("First 5 Rows")
display(df.head())
```

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumn Names:")
print(df.columns)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Dataset Information")
df.info()
```

```python
print(df.describe())
```

## Customer Churn Distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Churn", data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

## Monthly Charges Distribution

```python
plt.figure(figsize=(7,5))
sns.histplot(df["MonthlyCharges"], bins=20)
plt.title("Monthly Charges Distribution")
plt.show()
```

## Contract Type Distribution

```python
plt.figure(figsize=(7,5))
sns.countplot(x="Contract", data=df)
plt.title("Contract Type")
plt.xticks(rotation=20)
plt.show()
```

## Box Plot of Monthly Charges by Churn

```python
plt.figure(figsize=(7,5))
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

## Correlation Heatmap

```python
plt.figure(figsize=(8,6))

sns.heatmap(
    df.select_dtypes(include="number").corr(),
    annot=True,
    cmap="Blues"
)

plt.title("Correlation Heatmap")
plt.show()
```
